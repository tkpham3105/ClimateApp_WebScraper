```python
from facebook_scraper import get_posts # free of charge !!!
import pandas as pd
import json
from tqdm import tqdm
from datetime import datetime
```

```python
# Famous Facebook public groups on Climate Change news
GROUPS = ['NASAClimateChange', 'ClimateChangeHeadlines', 
         'climatenewsnetwork', 'CLEANET', 'Mason4C']

# Upper bound to scrape only up-to-date posts on each group 
MAX_POST = 200
```

```python
# Dataframe to store results
df = pd.DataFrame()

for i, group in enumerate(GROUPS):
    print("===================== Processing group number {} ======================".format(i+1))
    l_posts = []
    
    # Scrape until reach the upper bound
    for j, post in enumerate(get_posts(group, pages = 200)):
        if(j < MAX_POST):
            print("---------------------- Scraping post number {} ----------------------".format(j+1))
            l_posts.append(post)
        else:
            break
    df[group] = l_posts
    print('\n\n')
```

```python
# Filter post information about posted time (originally in datetime type), enabling df to be stored in .json format
def time_filtering(cell):
    try:
        cell['time'] = cell['time'].strftime("%m/%d/%Y, %H:%M:%S")
    except:
        return ''
    return cell

# tqdm helps observe progress of applying function on dataframe 
tqdm.pandas()

# Apply filtering function to each cell accordingly
df = df.progress_applymap(lambda x: time_filtering(x))
```

```python
df.to_json("res_facebook.json", orient = "columns")
```
